```python
from stanza.runtime import setup
setup()
```

```python
import stanza.env
import ipywidgets
envs = ["mujoco/robosuite/pickplace", "mujoco/robosuite/nutassembly", "mujoco/pusht"]

dropdown = ipywidgets.Dropdown(
    options=envs,
    value=envs[0],
    description='Environments:',
    disabled=False,
)
dropdown
```

```python
import jax
import jax.random
import jax.numpy as jnp

import stanza.env
from stanza.env import ImageRender

env = stanza.env.create(dropdown.value)
```

```python
from stanza.util.ipython import as_video, as_image
from stanza.canvas import image_grid
# generate a bunch of random intiailizations, render the initializations,
# and display in a grid
states = jax.vmap(env.reset)(jax.random.split(jax.random.key(43), 16))
images = jax.vmap(lambda s: env.render(s, ImageRender(256, 256)))(states)
as_image(image_grid(images))
```
