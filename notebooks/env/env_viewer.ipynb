```python
from stanza.runtime import setup
setup()
```

```python
import stanza.env
import ipywidgets
envs = ["mujoco/robosuite/can", "mujoco/pusht"]

dropdown = ipywidgets.Dropdown(
    options=envs,
    value=envs[0],
    description='Environments:',
    disabled=False,
)
dropdown
```

```python
import jax
import jax.random
import jax.numpy as jnp

import stanza.env
from stanza.env import ImageRender

env = stanza.env.create(dropdown.value)
```

```python
s = env.reset(jax.random.key(47))
imgs = []
for i in range(10):
    imgs.append(env.render(s,ImageRender(256, 256)))
    s = env.step(s, env.sample_action(jax.random.key(42)), jax.random.key(47))
```

```python
from stanza.util.ipython import as_video
as_video(jnp.stack(imgs))
```
