```python
from argon.runtime import setup
setup()
```

```python
import argon.env
import ipywidgets
envs = [
    "mujoco/robosuite/pickplace/random",
    "mujoco/robosuite/pickplace",
    "mujoco/robosuite/pickplace/can",
    "mujoco/robosuite/nutassembly",
    "mujoco/robosuite/nutassembly/random", 
    "mujoco/robosuite/nutassembly/round", 
    "mujoco/robosuite/nutassembly/square", 
    "mujoco/pusht"
]

dropdown = ipywidgets.Dropdown(
    options=envs,
    value=envs[0],
    description='Environments:',
    disabled=False,
)
dropdown
```

```python
import jax
import jax.random
import argon.numpy as jnp

import argon.env
from argon.env import ImageRender

env = argon.env.create(dropdown.value)
```

```python
from argon.util.ipython import as_video, as_image
from argon.canvas import image_grid
# generate a bunch of random intiailizations, render the initializations,
# and display in a grid
states = jax.vmap(env.reset)(jax.random.split(jax.random.key(42), 16))
images = jax.vmap(lambda s: env.render(s, ImageRender(256, 256)))(states)
as_image(image_grid(images))
```
